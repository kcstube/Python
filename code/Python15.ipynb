## break文

```python
count = 0
while True:
  print(count)
  count += 1
  if count == 5:
    break
```

```python
for i in range(5):
  print(i)
```

```python
nums = [2,4,6,8]
for j in nums:
  if j % 2 == 1:
    break
else:
  print("奇数の値を含めてください.")
```

## continue文

```python
for i in range(1,11):
  if i % 3 == 0:
    continue
  print(i)
```
