## Fizz Buzz

1から15までの数字を出力するプログラミングを書いてください。ただし、数字が3の倍数の時は数字の代わりにFizzと出力し、5の倍数の時は数字の代わりにBuzzと出力し、3と5の倍数のときは、FizzBuzzと出力すること。

```python
for i in range(1, 100):
  if i % 3 == 0 and i % 5 == 0:
    print('FizzBuzz')
  elif i % 3 == 0:
    print('Fizz')
  elif i % 5 == 0:
    print('Buzz')
  else:
    print(i)
```
